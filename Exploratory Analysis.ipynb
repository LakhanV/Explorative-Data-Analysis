```python
import pandas as pd
import random
import plotly.express as px


# Country information (Country, ISO Code, Region)
countries = [
    ("Germany", "DEU", "Europe"),
    ("France", "FRA", "Europe"),
    ("Spain", "ESP", "Europe"),
    ("India", "IND", "Asia"),
    ("China", "CHN", "Asia"),
    ("USA", "USA", "North America"),
    ("Brazil", "BRA", "South America"),
    ("Canada", "CAN", "North America"),
    ("Australia", "AUS", "Oceania"),
    ("Japan", "JPN", "Asia")
]

# Time period
months = ["Jan-2024", "Feb-2024", "Mar-2024"]

# Store generated records
data = []

# Generate sample renewable energy data
for country, iso, region in countries:
    for month in months:

        solar = random.randint(200, 600)
        wind = random.randint(150, 500)
        hydro = random.randint(100, 450)

        renewable_share = random.randint(30, 75)

        co2_reduction = random.randint(50, 300)

        gdp = random.randint(500, 25000)

        data.append([
            country,
            iso,
            region,
            month,
            solar,
            wind,
            hydro,
            renewable_share,
            co2_reduction,
            gdp
        ])

# Create DataFrame
df_geo = pd.DataFrame(data, columns=[
    "Country",
    "ISO_Code",
    "Region",
    "Month",
    "Solar_GWh",
    "Wind_GWh",
    "Hydro_GWh",
    "Renewable_%",
    "CO2_Reduction",
    "GDP_Billion"
])

# Save dataset
df_geo.to_csv("renewable_energy_geo_dataset.csv", index=False)

# Display first few records
print(df_geo.head())



# Use March 2024 data
df_map = df_geo[df_geo["Month"] == "Mar-2024"]

# Choropleth map showing renewable energy share
fig = px.choropleth(
    df_map,
    locations="ISO_Code",
    color="Renewable_%",
    hover_name="Country",
    color_continuous_scale="Greens",
    title="Renewable Energy Share by Country (March 2024)"
)

fig.show()


# Step 3: Solar Energy Production by Country


fig = px.bar(
    df_map.sort_values(by="Solar_GWh", ascending=False),
    x="Country",
    y="Solar_GWh",
    color="Region",
    hover_data=["Renewable_%"],
    title="Solar Energy Production by Country"
)

fig.show()


# Step 4: GDP vs Renewable Energy Share


fig = px.scatter(
    df_map,
    x="GDP_Billion",
    y="Renewable_%",
    size="Solar_GWh",
    color="Region",
    hover_name="Country",
    title="GDP vs Renewable Energy Share"
)

fig.show()


# Step 5: Renewable Energy Trends Over Time


fig = px.line(
    df_geo,
    x="Month",
    y="Renewable_%",
    color="Country",
    markers=True,
    title="Renewable Energy Share Over Time"
)

fig.show()
```
